### Dealing with Time-series Data ..is regression making sense?
The two main chalenges of time-series data is that 
- The different instances may-be correlated.Therefore,when we employ regressoin techniques,one if the fundamenatal assumptions that observations are independentor errors are uncorrelated does not hold true any more.
- In addition to this,there is trend and seasonality present in the data,Even if we incorporate these terms into our model,it is not a good idea to use regression on time-series data for forecasts

# BUILDING MODELS:

```python
#train,validations set split 
#last 4months for validation and the rest for training purpose
jet_tr = jet_trcopy.loc['2012-08-25':'2014-06-24']
jet_va = jet_trcopy.loc['2014-06-25':'2014-09-25']
```

```python
print(jet_va.shape)
print(jet_tr.shape)
```

```python
#plotting the count for both train and validations set on the same graph 
jet_tr.Count.plot(figsize=(12,7),label = "train-set foot-traffic count",color="blue")
jet_va.Count.plot(figsize = (12,7),label ="vaildation-set foot-traffic count",color = "orange")
plt.xlabel("time")
plt.ylabel('Count')
plt.legend(loc= "upper left")
plt.show()
```

The blue color is the timeseries data for training and the orange is for validation purpose,we will strictly perform traning only on training set and keep validatio set to validate our models<br>

# MODEL:1 NAIVE APPROACH
We assume that the next time-series data is the same as its previous data.That means,our model is  a constant not depending on anny of the feature

```python
from sklearn.metrics import mean_squared_error
import math
jet_tr_array = np.asarray(jet_tr.Count)
y_pred = jet_va.copy()
y_pred["naive"] = jet_tr_array[len(jet_tr_array)-1]
plt.figure(figsize=(12,7))
plt.plot(jet_tr.index,jet_tr.Count,label="train-count",color = "lightblue")
plt.plot(jet_va.index,jet_va.Count,label="validation-count")
plt.plot(y_pred.index,y_pred["naive"],label = "forecast on validation set",color = "red")
plt.legend(loc="upper left")
plt.show()
rms_naive = math.sqrt(mean_squared_error(jet_va.Count,y_pred.naive))
print("The RMS for naive approach is:",rms_naive)
```

MEH!!! As expected,we have very high RMSE.Let us build some complez models

# MODEL:2 MA(p) MODEL 
In moving average,we take the average of previous few datapoints to predict the next value in the time-series.The paramter - p in this process depicts the number of previous data-points we consider for building our model.We will build a MA(5),MA(10),MA(25),MA(50) and compare their performance on forecasating

```python
p = [5,10,25,50]
for i in p:
    y_pred_MA = jet_va.copy()
    y_pred_MA['MA_forecast'] = jet_tr['Count'].rolling(i).mean().iloc[-1]
    plt.figure(figsize=(12,7))
    plt.plot(jet_tr["Count"],label="train-count",color = "lightblue")
    plt.plot(jet_va["Count"],label="validation-count")
    plt.plot(y_pred_MA["MA_forecast"],label = "forecast on validation set",color = "red")
    plt.legend(loc="upper left")
    plt.show()
    rms_MA = math.sqrt(mean_squared_error(jet_va.Count,y_pred_MA["MA_forecast"]))
    print("The RMS for MA approach is:",rms_MA)
```

#### OBSERVATIONS:
- We see as p increases,the model's accuracy on vaidaiton set deccreases.


# MODEL:3 - EXPONENTIAL SMOOTHING
This model works similar to the MA model,except for the fact that the new data is found by taking **weighted average** over all the past data and the weights decrease exponentially as the time from the present data point increases(ie as we move further into the past).This means ,we give more weightage to the recent vlaues in preidicting a point

```python
from statsmodels.tsa.api import ExponentialSmoothing,SimpleExpSmoothing,Holt
levels = [0.2,0.4,0.6,0.8,1]
for i in levels:
    y_pred_SEM = jet_va.copy()
    fit2 = SimpleExpSmoothing(np.asarray(jet_tr["Count"])).fit(smoothing_level = i,optimized=False)
    y_pred_SEM["sem"] = fit2.forecast(len(jet_va))
    plt.figure(figsize=(12,7))
    plt.plot(jet_tr["Count"],label="train-count",color = "lightblue")
    plt.plot(jet_va["Count"],label="validation-count")
    plt.plot(y_pred_SEM["sem"],label = "forecast on validation set",color = "red")
    plt.legend(loc="upper left")
    plt.show()
    rms_SEM = math.sqrt(mean_squared_error(jet_va.Count,y_pred_SEM["sem"]))
    print("The RMS for MA approach is:",rms_SEM)
```

#### OBSERVATIONS:
- As we increase the level of smoothing,the model performs better
- The model performs better than MA and naive approach

# MODEL:4 - HOLT'S LINEAR TREND MODEL
Time has come,that we have to acknowledge the presence of trend in the data and incorporate a model which can handle this efficiently.That is why we introduce holt's model which takes care of both the level as well as the trend in the data.This is basically a extension of exponential smoothing with additional benefit of trend in the model

```python
import statsmodels.api as sm
sm.tsa.seasonal_decompose(jet_tr.Count).plot()
res = sm.tsa.stattools.adfuller(jet_tr.Count)
plt.show()
```

```python
y_hat_avg = jet_va.copy()
fit1 = Holt(np.asarray(jet_tr['Count'])).fit(smoothing_level= 0.3,smoothing_slope = 0.07)
y_hat_avg["holt"]=fit1.forecast(len(jet_va))
plt.figure(figsize=(12,7))
plt.plot(jet_tr["Count"],label="train-count",color = "lightblue")
plt.plot(jet_va["Count"],label="validation-count")
plt.plot(y_hat_avg["holt"],label = "forecast on validation set",color = "red")
plt.legend(loc="upper left")
plt.show()
rms_SEM = math.sqrt(mean_squared_error(jet_va.Count,y_hat_avg["holt"]))
print("The RMS for MA approach is:",rms_SEM)
```

#### OBSERVATIONS:
-We can see the inclined line here as the model has taken into consideration the trend present in the data
